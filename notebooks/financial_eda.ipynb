```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned_financial_data.csv")
df.head()
df.info()
df.describe()

df.groupby('segment')['profit'].sum().sort_values().plot(kind='barh')
plt.title("Total Profit by Segment")
plt.show()

# Percentage of high-risk vs low-risk
df['risk_flag'].value_counts(normalize=True) * 100

# High-risk count by segment
df[df['risk_flag'] == 'High Risk'].groupby('segment').size().sort_values(ascending=False)

#Compare profit
df.groupby('risk_flag')['profit'].mean()
```

- Most records fall under Low Risk, indicating overall stable profitability.

- High Risk entries have low profit margins and contribute less to total profit.

- High Risk records are concentrated in specific business segments.

- Low Risk records generate significantly higher average profit.

- Improving margins in High Risk segments can materially improve overall performance.
